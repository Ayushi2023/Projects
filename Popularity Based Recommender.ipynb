```python
movie = pd.read_csv("Datasets/movies.csv")
rating = pd.read_csv("Datasets/ratings.csv")
```

```python
rating.head()
```

```python
df = pd.merge(movie,rating, on="movieId")
df.head()
```

```python
record = pd.DataFrame(df.groupby(["title","genres"])["rating"].mean())
```

```python
record["Num Reviews"] = pd.DataFrame(df.groupby(["title","genres"])["userId"].count())
```

```python
record
```

```python
record = record.reset_index()
```

```python
record.columns = ["Movie Title","Genre","Average Movie Ratings","Num Reviews"]
```

```python
g = input("Enter Genre: ")
t = int(input("Enter threshold value : "))
N = int(input("Enter num of recommedations : "))

result = record.sort_values("Average Movie Ratings",ascending=False)[record["Num Reviews"] >= t]
result[result["Genre"] == g].head(N)
```
